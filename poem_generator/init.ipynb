```python
def decode(c):
    if c == 1:
        return 9
    elif c == 3:
        return 13
    elif c == 127 - 30:
        return 10
    elif 32 <= c + 30 <= 126:
        return c + 30
    else:
        return 0

def encode(c):
    # '\t'
    if c == 9:
        return 1
    # '\r'
    elif c == 13:
        return 3
    # '\n'
    elif c == 10:
        return 127 - 30
    elif 32 <= c <= 126:
        return c - 30
    else:
        return 0
```
