```python
import os
import h5py
from matplotlib import pyplot as plt
import random
```

```python
df =  h5py.File('post_trainset/dataset.h5', "r")

source_list = df['source'][:]

destination_list = df['destination'][:]

left_names = df['left_names'][:]

right_names = df['right_names'][:]

pair_list = df['pair'][:]

print( 'source : ' ,  source_list.shape )

print( 'destination : ' ,  destination_list.shape )
```

```python
fig = plt.figure()

idxs = random.sample( range( 0 , len(source_list) ) , 4 )

for i in range(len(source_list)):
    
    idx = idxs[i]
    
    image = source_list[ idx ]

    ax = plt.subplot(1, 4 , i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{}'.format(idx))
    ax.axis('off')
    plt.imshow(image)
    # show_landmarks(**sample)

    if i == 3:
        plt.show()
        break
```
